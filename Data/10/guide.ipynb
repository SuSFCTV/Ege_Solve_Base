```python
f = open("data/K2033/2033.txt")

sub_str = "ее"
ans = []
strange_symbols = [",", ".", "?", ")", "(", "-", "!","[", "]", ";", "'", '"', ":"]
for line in f:
    for elem in strange_symbols:
        line = line.replace(elem, "")
        print(line)
        break
    for word in line.split():
        if word.find(sub_str) != -1 and len(word) > 2:
            ans.append(word)
            
print(len(ans))
```
