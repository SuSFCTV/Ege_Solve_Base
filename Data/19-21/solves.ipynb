### Task - 3075

![](https://github.com/vaskers5/Ege_Solve_Base/blob/main/Data/19-21/data/3075.png?raw=true)

```python
from functools import lru_cache

def moves(h):
    return h * 2, h + 2

@lru_cache(None)
def f(h):
    if h>=25:
        return "СР"
    elif any(f(m) =="СР" for m in moves(h)):
        return "П1"
    elif all(f(m) == "П1" for m in moves(h)):
        return "В1"
    elif any(f(m) == "В1" for m in moves(h)):
        return "П2"
    elif all(f(m) == "П1" or f(m)=="П2" for m in moves(h)):
        return "В2"
    
for i in range(1,30):
    print(i, ":", f(i))
```

```python
from functools import lru_cache

def moves(h):
    return h * 3, h * 2, h + 1

@lru_cache(None)
def f(h):
    if 36<=h<=60:
        return "СР"
    elif h>60:
        return "П1"
    elif any(f(m) =="СР" for m in moves(h)):
        return "П1"
    elif all(f(m) == "П1" for m in moves(h)):
        return "В1"
    elif any(f(m) == "В1" for m in moves(h)):
        return "П2"
    elif all(f(m) == "П1" or f(m)=="П2" for m in moves(h)):
        return "В2"
    
for s in range(1,35):
    print(s,":",f(s))
```

### Task - 3190

```python
from functools import lru_cache

def moves(h):
    a, b = h
    return (a + 1, b), (a, b + 1), (a * 4, b), (a, b * 4)

@lru_cache(None)
def f(h):
    if sum(h) >= 125:
        return "СР"
    elif any(f(m) =="СР" for m in moves(h)):
        return "П1"
    elif all(f(m) == "П1" for m in moves(h)):
        return "В1"
    elif any(f(m) == "В1" for m in moves(h)):
        return "П2"
    elif all(f(m) == "П1" or f(m)=="П2" for m in moves(h)):
        return "В2"
    
for s in range(1,118):
    print(s,":",f( (7,s) ) )
```

### Task - 3489

![](https://github.com/vaskers5/Ege_Solve_Base/blob/main/Data/19-21/data/3489.png?raw=true)

```python
from functools import lru_cache

def moves(h):
    a,b = h
    if a < 2:
        return (a, b // 2), (a, b - 1)
    elif b < 2:
        return (a // 2, b), (a - 1, b)
    elif a > 1 and b > 1:
        return (a, b // 2), (a, b - 1), (a // 2, b), (a - 1, b)

@lru_cache(None)
def f(h):
    if sum(h) <= 18:
        return "СР"
    elif any(f(m) =="СР" for m in moves(h)):
        return "П1"
    elif all(f(m) == "П1" for m in moves(h)):
        return "В1"
    elif any(f(m) == "В1" for m in moves(h)):
        return "П2"
    elif all(f(m) == "П1" or f(m)=="П2" for m in moves(h)):
        return "В2"
    
for m in range(1, 100):
    if f((m, m)) == "В1":
        print(m)
print()

k = 13
for s in range(1, 100):
    if f((k, s)) == "П2":
        print(s)
print()
for n in range(1, 100):
    if f((n, n)) == "В2":
        print(n)
```
